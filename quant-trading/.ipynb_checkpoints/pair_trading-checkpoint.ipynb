# Pair Trading

## Introduction

We will build a mean reverting portfolio with multiple companies and perform backtest. Here is the plan:

1. Start with top 10 energy companies.
2. First, for each pair, perform the ADF test to find the most cointegrated pair.
3. For the chosen pair, compute the PnL and the Sharpe ratio to see whether the strategy is profitable.
4. Apply the Johansen test and find the weights that provide the cointegrated portfolio from the 10 tickers.
5. Perform the backtest to see if the plan is profitable.

### ADF Test

For each pair among the companies, we perform ADF test to find the most cointegrated pair.

```python
# ADF for all combinations of two stocks
results_adf = []

for x, y in itertools.combinations(tickers, 2):
    # estimate hedge ratio beta
    X    = sm.add_constant(logp[x])
    y_s  = logp[y]
    ols  = sm.OLS(y_s, X).fit()
    beta    = ols.params.iloc[1]            
    resid = y_s - (ols.params.iloc[0] + beta * logp[x])
    
    # ADF test on residuals
    adf_res = adfuller(resid, maxlag=1, regression='c')
    pvalue  = adf_res[1]
    
    # half‐life
    eps_lag = resid.shift(1).dropna()
    delta   = resid.diff().dropna()
    armod   = sm.OLS(delta.loc[eps_lag.index], sm.add_constant(eps_lag)).fit()
    phi       = armod.params.iloc[1] + 1      # again use .iloc
    half_life = -np.log(2) / np.log(phi) if phi > 0 else np.nan
    
    # spread volatility
    spread_vol = resid.std()
    
    results_adf.append({
        'pair':       (x, y),
        'pvalue':     pvalue,
        'half_life':  half_life,
        'volatility': spread_vol
    })

# print result
df_adf = pd.DataFrame(results_adf).sort_values(
    ['pvalue', 'half_life', 'volatility'],
    ascending=[True, True, True]
).reset_index(drop=True)

print(df_adf)
```

To compute trading signals and backtest for a mean-reverting strategy, we define a function as follows:

Parameters:
- spread (pd.Series): stationary spread series
- window (int): rolling window length for mean/std
- high_z (float): z-score threshold to enter short
- low_z (float): z-score threshold to enter long
- close_z (float): z-score threshold to exit

Returns:
- pnl (pd.Series): daily PnL series
- cum_return (pd.Series): cumulative PnL
- sharpe (float): annualized Sharpe ratio
- drawdown (pd.Series): drawdown series
- max_dd (float): maximum drawdown

```python
def mean_reverting(spread, window, high_z, low_z, close_z):

    # Rolling statistics
    mu    = spread.rolling(window).mean()
    sigma = spread.rolling(window).std()
    zscore = (spread - mu) / sigma

    # Generate signals
    signals = pd.Series(0.0, index=spread.index)
    signals[zscore >  high_z] = -1.0
    signals[zscore < -low_z] = +1.0
    signals[zscore.abs() < close_z] = 0.0
    signals = signals.ffill().fillna(0.0)

    # Compute PnL
    delta_spread = spread.diff()
    pnl          = signals.shift(1) * delta_spread

    # Performance metrics
    cum_return = pnl.cumsum()
    sharpe     = pnl.mean() / pnl.std() * np.sqrt(252)
    drawdown   = cum_return - cum_return.cummax()
    max_dd     = drawdown.min()

    # Plot cumulative return
    fig, ax = plt.subplots()
    ax.plot(cum_return.index, cum_return)
    ax.set_xlabel('Date')
    ax.set_ylabel('Cumulative PnL')
    ax.set_title(f'Cumulative PnL\nSharpe: {sharpe:.2f}, Max DD: {max_dd:.2f}')
    plt.tight_layout()
    plt.show()
    
    # Plot drawdown
    fig, ax = plt.subplots()
    ax.plot(drawdown.index, drawdown)
    ax.set_xlabel('Date')
    ax.set_ylabel('Drawdown')
    ax.set_title('Drawdown Over Time')
    plt.tight_layout()
    plt.show()
    # Store results
    return pnl, cum_return, sharpe, drawdown, max_dd

#def plot_mean_reverting(pnl, cum_return, sharpe, drawdown, max_dd):
    
   
    
```

### Johansen test

```python
jres = coint_johansen(logp, det_order=0, k_ar_diff=1)
```

```python
# Trace Stats
trace_stats = jres.lr1
crit_90    = jres.cvt[:, 0]
crit_95    = jres.cvt[:, 1]
crit_99    = jres.cvt[:, 2]
ranks      = list(range(len(trace_stats)))

results_trace = pd.DataFrame({
    'r (≤ # coint rel)': ranks,
    'Trace Stat':         trace_stats,
    'CV 90%':             crit_90,
    'CV 95%':             crit_95,
    'CV 99%':             crit_99,
})
results_trace['Reject @95%'] = results_trace['Trace Stat'] > results_trace['CV 95%']

print(results_trace)

fig, ax = plt.subplots()
x = np.arange(1, len(trace_stats)+1)
ax.bar(x, trace_stats)
ax.plot(x, crit95, marker='x', linestyle='--', label='95% crit. value')
ax.set_xlabel('Hypothesized # of Coint. Relations')
ax.set_ylabel('Trace Statistic')
ax.set_title('Johansen Trace Test')
ax.legend()
plt.show()
```

The result says that with 95% critical level, we cannot reject the null hypothesis that there is at least one cointegration pair. However, we are still able to proceed with the first eigenvector to find the cointegrated portfolio.

```python
# The eigenvalues
eigvals = jres.eig
fig, ax = plt.subplots()
ax.plot(np.arange(1, len(eigvals)+1), eigvals, marker='o')
ax.set_xlabel('Index')
ax.set_ylabel('Eigenvalues')
ax.set_title('Johansen Eigenvalues')
plt.show()
```

```python
# Weights from the first eigen vector
first_evec = jres.evec[:, 0]

strong_series = pd.Series(first_evec, index=tickers)
weights = strong_series / strong_series.sum()

# 5) Plot the weights
fig, ax = plt.subplots()
ax.bar(weights.index, weights.values)
ax.set_xlabel('Ticker')
ax.set_ylabel('Portfolio Weight')
ax.set_title('Normalized Cointegrating‐Vector Weights (sum=1)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
spread_jhs    = logp.dot(weights)
fig, ax = plt.subplots()
ax.plot(spread_jhs.index, spread_jhs)
ax.set_xlabel('Date')
ax.set_ylabel('Spread')
ax.set_title('Time Series of the Cointegrated Spread')
plt.tight_layout()
plt.show()
```
